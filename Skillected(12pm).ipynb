```python
# break :- break is use toterminate the loop immediately.
i=1
while i<=10:
  if i==5:
    break
  print(i)
  i+=1
```

```python
# Continue :- its skip the corrent iteration and move ]s to the next iteration.
i=0
while i<5:
  i+=1

  if i==3:
    continue
  print(i)
```

```python
secret=7
attempts=0
while True:
  guess=int(input("guess the number "))
  attempts+=1
  if guess==secret:
    print("correct! you won ",attempts, "attempts")
    break
  elif guess<secret:
    print("too low")
  elif guess>secret:
    print("too high")
```
