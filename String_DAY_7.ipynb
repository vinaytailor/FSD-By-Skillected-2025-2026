```python
# reverse string

word = input("enter: ")
rev = word[::-1]
print("reverse:", rev)
```

```python
# palindrome

word = input("enter: ")

if word == word[::-1]:
    print("Palindrome")
else:
    print("Not palindrome")
```

```python
# email splitting

email = input("enter email: ")

parts = email.split("@")

print("username:", parts[0])
print("domain:", parts[1])
```

```python
# password validation

password = input("enter password: ")

if len(password) >= 8:
    print("Valid password")
else:
    print("Invalid password (min 8 characters)")
```
